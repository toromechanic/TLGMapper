# Enterprise Compliance Automation for TLGMapper

This notebook inspects the repository state, applies compliance updates, validates the repository, and prepares pull request metadata for an enterprise-ready change set.

```python
import os
import subprocess
from pathlib import Path

ROOT = Path('/workspaces/TLGMapper')
print('Repository root:', ROOT)
print('Exists:', ROOT.exists())
print('Git available:', subprocess.run(['git', '--version'], capture_output=True, text=True).returncode == 0)
print('Files under root:')
for path in sorted(ROOT.glob('*')):
    print('-', path.name)
```

## Identify Compliance Requirements

This repository requires enterprise compliance coverage for:

- supply chain transparency and dependency provenance
- security policy documentation
- CI validation for repository artifacts
- audit metadata for generated outputs
- ownership metadata for review and governance


```python
import subprocess
from pathlib import Path

print('Changed files:')
result = subprocess.run(['git', 'status', '--short'], cwd=Path('/workspaces/TLGMapper'), capture_output=True, text=True)
print(result.stdout)

print('Git diff summary:')
result = subprocess.run(['git', 'diff', '--name-only'], cwd=Path('/workspaces/TLGMapper'), capture_output=True, text=True)
print(result.stdout)
```

## Update Policy and Configuration Files

The repository has been updated with the following compliance assets:

- `.gitignore`
- `.github/workflows/ci.yml`
- `.github/CODEOWNERS`
- `SECURITY.md`
- `ASSET_PROVENANCE.md`
- `DEPENDENCIES.md`
- `scripts/validate.py`
- `README.md` compliance section additions


```python
import py_compile
from pathlib import Path

sources = [
    Path('/workspaces/TLGMapper/TLGMapper.py'),
    Path('/workspaces/TLGMapper/scripts/validate.py'),
]

success = True
for source in sources:
    try:
        py_compile.compile(source, doraise=True)
        print(f'Compiled successfully: {source.name}')
    except Exception as exc:
        print(f'Validation failed for {source.name}: {exc}')
        success = False

print('GitHub workflow exists:', Path('/workspaces/TLGMapper/.github/workflows/ci.yml').exists())
print('Security doc exists:', Path('/workspaces/TLGMapper/SECURITY.md').exists())
print('Dependency doc exists:', Path('/workspaces/TLGMapper/DEPENDENCIES.md').exists())
print('Asset provenance doc exists:', Path('/workspaces/TLGMapper/ASSET_PROVENANCE.md').exists())
print('Validation status:', 'PASS' if success else 'FAIL')
```

## Prepare Pull Request Metadata

The following PR metadata is generated for the compliance update branch.


```python
import subprocess
from pathlib import Path

repo = Path('/workspaces/TLGMapper')
changed = subprocess.run(['git', 'diff', '--name-only'], cwd=repo, capture_output=True, text=True).stdout.strip().splitlines()

pr_title = 'Enterprise compliance: provenance, security docs, validation workflow'
pr_description = '''\
## Enterprise Compliance Enhancements

This PR upgrades `TLGMapper` with enterprise-grade compliance controls:

- added CI validation workflow for Python syntax and repository integrity
- added `SECURITY.md`, `DEPENDENCIES.md`, and `ASSET_PROVENANCE.md`
- added ownership metadata via `.github/CODEOWNERS`
- added `.gitignore` to exclude bytecode artifacts
- added `scripts/validate.py` for local repository validation
- enhanced `README.md` with compliance and provenance guidance

## Changed files
'''
for path in changed:
    pr_description += f'- {path}\n'

print('PR title:')
print(pr_title)
print('\nPR description:')
print(pr_description)
```
